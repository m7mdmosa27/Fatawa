```python

import re
from pyarabic.araby import strip_tashkeel
from pyarabic.araby import tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('arabic'))
stop_words.add('وعلى')
def clean_text_arabic(text):
    # Remove sequences of the form "[number:arabic_word]"
    text = re.sub(r'\[[\u0600-\u06FF]+\s*:*\s*\d*\]', '', text)
    # Find URLs and email addresses in the text
    text = re.sub(r'\b(?:https?|ftp)://\S+|www\.\S+|\S+@\S+|\b[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}\b/*\S*', '', text)
    # Remove special characters and punctuation
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'_+', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Tokenization
    words = tokenize(text)
    # Remove diacritics
    words = [strip_tashkeel(word) for word in words]
    # Remove stopwords
    cleaned_text = [word for word in words if word not in stop_words]
    # Join words
    cleaned_text = ' '.join(cleaned_text)
    cleaned_text = cleaned_text.replace('الحمد لله والصلاة والسلام رسول الله وآله وصحبه', "")
    cleaned_text = cleaned_text.replace('الحمد لله والصلاة والسلام رسول الله آله وصحبه', "")
    return cleaned_text

# Example
text = "هذا هو نَصُ ٧٥٤٧٥٥ تجرُيبٌي! يحتٍوْي على بعض  5151 العلامات  ٧٦' [الإنسان: 7]' التجريبية؟"
text = """الحمد لله والصلاة والسلام على رسول الله وآله وصحبه أما بعد:
نعم يمكن للأخت أن ترضع أخاها ولو لم تكن الأم مريضة ويترتب على رضاعها لأخيها ما يترتب عليه من الاحكام ، فقد صار ابناً لها وابنا لزوجها الذي نشأ اللبن عن وطئه، فلا يمكن أن يتزوج ابنة هذا الرجل ولو كانت من غير أخته التي أرضعته، ولا يتزوج أخت هذا الرجل ولا خالته ولا عمته لأنه صار أباه. أما الدليل فهو قول الله جل وعلا: حرمت عليكم أمهاتكم وبناتكم  الآيه [النساء: 23]. وقول النبي صلى الله عليه وسلم : يحرم من الرضاع ما يحرم من النسب .5151  [ أخرجه البخاري وغيره ]. والله أعلم.
الحمد لله والصلاة والسلام على رسول الله وعلى آله وصحبه وسلم ..... وبعد:             إذا دعت الضرورة إلى التحويل عن طريق البنوك الربوية، فلا حرج في ذلك إن شاء الله، لقوله تعالى: (وقد فصل لكم ما حرم عليكم إلا ما اضطرتم إليه). الأنعام:119.       ولا شك أن التحويل عن طريقها من الضرورات العامة في هذا العصر، وهكذا الإيداع فيها للضرورة بدون اشتراط فائدة، فإن دفعت إليك الفائدة من دون اشتراط ولا اتفاق فلا بأس بأخذها لصرفها في المشاريع الخيرية كمساعدة الفقراء والغرماء ونحو ذلك. لا لتتملكها أو تنتفع بها، بل هي في حكم المال الذي يضر تركه بالمسلمين مع كونه من مكسب غير جائز، فصرفه فيما ينفع ٧٥٤٧٥٥ المسلمين أولى من تركه لمن يستعينون به على ما حرم الله.    
http://www.almslm.com/bitalnbibaitalnabihtm mohamed@gmail.com  facebock.com/mohamed
   وإن أمكنه الإيداع والتحويل عن طريق بنوك إسلامية أو طرق مباحة أخرى، لم يجز الإيداع في البنوك  _________ الربوية ___لزوال الضرورة والله أعلم
الحمد لله والصلاة والسلام على رسول الله وعلى آله وصحبه، أما بعد:"""
cleaned_text = clean_text_arabic(text)
print(cleaned_text)
```

```python
def is_single_word(text):
    words = text.split()
    return len(words) <= 1

print(is_single_word(' '))
print(is_single_word(' والله '))
print(is_single_word(' والله العظيم'))
```

```python
def tokenize_dataset(df):
    dataset = df['ans_cleaned'].to_list() + df['title_cleaned'].to_list() + df['ques_cleaned'].to_list()
    splited_dataset = [text.split('\n') for text in dataset]
    tokenized_dataset=[]
    count = 0
    for text_list in splited_dataset:
        count +=1
        for sequance in text_list:
            if not is_single_word(sequance):
                tokenized_dataset.append(sequance.split())
    print(len(tokenized_dataset))
    print(count)
    return tokenized_dataset
```

```python
df_tokenized = tokenize_dataset(df)
```

```python
def get_word_vectors(sentence, model):
    word_vectors = []
    for word in sentence.split():
        try:
            word_vectors.append(model.wv[word])
        except KeyError:
            # If the word is not in the vocabulary, skip it
            continue
    return word_vectors

def get_sentence_vector(sentence_vectors):
    if len(sentence_vectors) > 0:
        return np.mean(sentence_vectors, axis=0)
    else:
        return np.zeros((150,))
```

```python
def sent2vec(user_question, model):
    user_question = clean_text_arabic(user_question)
    user_question = get_word_vectors(user_question, model)

    return get_sentence_vector(user_question)
```

```python
def get_most_similar_question(user_question, dataset_questions):
    # Preprocess the user question
    user_question = clean_text_arabic(user_question)
    df = dataset_questions
    # Preprocess dataset questions
    dataset_questions = [clean_text_arabic(q) for q in dataset_questions]
    
    # Initialize TF-IDF vectorizer
    vectorizer = TfidfVectorizer(stop_words=list(stop_words), analyzer= 'char_wb')
    
    # Fit and transform the dataset questions
    X = vectorizer.fit_transform(dataset_questions)
    
    # Transform the user question
    user_question_vec = vectorizer.transform([user_question])
    
    # Calculate cosine similarity between user question and dataset questions
    similarity_scores = cosine_similarity(user_question_vec, X)
    print(similarity_scores)
    
    # Get the index of the most similar question
    most_similar_index = similarity_scores.argmax()
    print(user_question)
    print(most_similar_index)
    # Return the index of the most similar question
    return df.loc[most_similar_index]
```
